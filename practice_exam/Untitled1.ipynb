```python
def modify(L, D):
    ret = []
    for l in L:
        if l in D:
            ret.append(D[l])
        else:
            ret.append(l)
    return ret

def in_base(n, b):
    try:
        return int(n, b)
    except:
        return None

def basegroup(x):
    ret = {}
    for num in x:
        for i in range(2, 11):
            new_num = in_base(num, i)
            if new_num != None:
                if new_num in ret:
                    if num not in ret[new_num]:
                        ret[new_num].append(num)
                else:
                    ret[new_num] = [num]
    new_ret = {}
    for key in ret:
        if len(ret[key]) > 1:
            new_ret[key] = ret[key]
    return new_ret
              
class Point():
    
    def __init__(self, x=None, y=None):
        self.x = x
        self.y = y
        
class Circle():
    
    def __init__(self, x, y, radius):
        self.x = x
        self.y = y
        self.radius = radius
        
    def __contains__(self, point):
        if point.x == None:
            return False
        dx = (self.x - point.x)**2
        if point.y == None:
            return False
        dy = (self.y - point.y)**2
        return self.radius > (dx+dy)**0.5
        
def main():
    # modify testing
    print('modify tests')
    L=['a','b',2,'a','c']
    D={2:"two",'a':1}
    modify(L,D)
    print('modify',L)
    L=[6, 8, 12, 6, 12,'alpha']
    D={6:"two",'alpha':42}
    modify(L,D)
    print('modify',L)
    # basegroup testing
    print('\nbasegroup tests')
    for grouptest in [ ['3','4','21','7'], ['100','4','5'], ['6','7','21']]:
      res = basegroup(grouptest)
      print(grouptest,res)

    # Circle and Point testing
    print('\nCircle/Point tests')
    points = [Point(4,6),Point(5,8),Point(4,8),Point(6.1,8),Point()]
    
    r = Circle(4,6,2.1)

    print('testing',points,'with',r)
    for q in points:
        if q in r:
            print(q ,'is inside of',r)
        else:
            print(q,'is outside of',r)
    print([r])
    
main()
```
